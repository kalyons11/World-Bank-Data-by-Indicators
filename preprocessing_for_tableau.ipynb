```python
import pandas as pd
```

```python
keys = ['climate-change', 'energy-and-mining', 'economy-and-growth']
```

```python
with open('bad_countries.txt', 'r') as f:
    bad = list(map(lambda s: s.strip(), f.readlines()))
```

```python
bad[:5]
```

```python
data_filt_bad = dict()
for key in keys:
    data = pd.read_csv(f'{key}/{key}.csv')
    data_filt = data[~(data['Country Name'].isin(bad))]
    data_filt_bad[key] = data_filt
    print(f'{key} complete - filtered bad country names out.')
```

```python
min_year, max_year = 1970, 2016
```

```python
data_filt_year = dict()
for key in keys:
    data = data_filt_bad[key]
    data_filt = data[(data.Year >= min_year) & (data.Year <= max_year)]
    data_filt_year[key] = data_filt
    print(f'{key} complete - filtered partially missing years out.')
```

```python
bad_years = [1970, 1980, 1990]
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
```

```python
data_filt_double_counting = dict()
for key in keys:
    data = data_filt_year[key]
    data_years = data[data.Year.isin(bad_years)]
    print(data_years.dtypes)
    test = data_years.select_dtypes(include=numerics)
    test /= 2
    data_years.loc[:, list(set(test.columns) - {'Year'})] = test
    # print(data_years[data_years['Country Name'] == 'United States'])
    data.loc[data_years.index, :] = data_years
    data_filt_double_counting[key] = data
    print(f'{key} complete - filtered bad double-counting years out.')
```

```python
for key in keys:
    data = data_filt_double_counting[key]
    data.to_csv(f'{key}/{key}-filt.csv', index=None)
    print(f'successfully filtered and exported {key}')
```
